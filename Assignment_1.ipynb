1. Why are dictionary keys required to be immutable in Python? Explain with a small example.

In dictionary data is stored in key value pairs. Key is used to quickly find the value. if a key is changed after being stored it will not able to find its value again
that's the dictionary key is immutable and cannot be change after creation.
For example- 


```python
# string and tuple as key (immutable)
student = {
    "name": "Diksha",
    (1, 2): "Roll numbers"
}

print(student)
```

```python
# list as key (mutable)
stud = {
    [1, 2, 3]: "Marks"
}
```

2. Write a program that takes a list of elements and creates a dictionary showing the frequency of each element.

```python
lst= [1, 2, 2, 3, 4, 1, 2, 3]
feq = {}
for item in lst:
    feq[item] = feq.get(item,0) + 1
print("Frequency of each element:")
print(feq)
```

3. What is the difference between dict.get(key) and dict[key]? Write a code example where get() is safer.

dict[key] raises a KeyError if the key does not exist.
dict.get(key) returns None or a default value if the key is missing.
Hence, get() is safer to use.
For Example-

```python
student = {
    "name": "Diksha",
    "age": 20
}
print(student["name"])   
```

```python
print(student["marks"])  # it gives error bcz markes key does not exist
```

```python
print(student.get("marks", 0))    # set default value 0 so does not return error 
```

4. Write a program to merge two dictionaries and resolve common keys by keeping the larger value.

```python
dict1 = {"a": 10, "b": 20, "c": 30}
dict2 = {"b": 40, "c": 50, "d": 60}
merged_dict = {}
for key in dict1.keys() | dict2.keys():
    merged_dict[key] = max(dict1.get(key, 0), dict2.get(key, 0))
print("Merged Dictionary:")
print(merged_dict)
```

5. Using a dictionary comprehension, create a dictionary from a list of integers where keys are numbers and values are their cubes, but only for odd numbers.

```python
lst = [1, 2, 3, 4, 5, 6, 7]

cube_dct = {num: num**3 for num in lst if num % 2 != 0}
print("Dictionary of odd number with its cubes:")
print(cube_dct)
```
